```python
import pandas as pd
import numpy as np
import citipy.citipy as citipy
import requests
import json
import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import urllib
```

```python
cities = []
country = []

#Floats (More Exact)
lats = np.random.uniform(low=-90, high=90, size=(500,))
lons = np.random.uniform(low=-180, high=180, size=(500,))
# The code for random Integers    
#lats = np.random.randint(low=-90, high=90, size=500)
#lons = np.random.randint(low=-180, high=180, size=500)

#we join them with a zip and we do the nearest city search with citipy
for lat, lon in zip(lats, lons):
    print(lat, lon)    
    city = citipy.nearest_city(lat, lon)
    print(city)
```

```python
for lat, lon in zip(lats, lons):
    print(lat, lon)    
    city = citipy.nearest_city(lat, lon)
    city= city.city_name
    cities.append(city)
    print(city)

cities.head()
```

```python
# Weather
api_key = "9bebaf9cb5a86a22e739abe823773f28"

# Endpoint URL for five day forecast Phoenix, AZ
target_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

city_data= []

#API call by lat/lon, city name and cityname + country code
#api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}
#api.openweathermap.org/data/2.5/weather?q={city name}
#api.openweathermap.org/data/2.5/weather?q={city name},{country code}

for i, city in enumerate(cities):
    print(city)
    city_url = target_url+"&q=" + urllib.request.pathname2url(city)
 
    try:
        city_weather=requests.get(city_url).json()
        
        city_lat = city_weather["coord"]["lat"]
        city_lon = city_weather["coord"]["lon"]
        city_clouds = city_weather["clouds"]["all"]
        city_country = city_weather["sys"]["country"]
        city_hum = city_weather["main"]["humidity"]
        city_max = city_weather["main"]["temp_max"]
        city_wind = city_weather["wind"]["speed"]
        
        city_data.append({"City":city, "Lat":city_lat, "Lon":city_lon, "Cloudiness":city_clouds, "Country":city_country, 
                          "Humidity":city_hum, "Max Temp":city_max, "Wind Speed":city_wind})
        
        #City    Cloudiness    Country    Date    Humidity    Lat    Lng    Max Temp    Wind Speed
        
        print(city_weather)
        
    except: 
        print("City not found")
        pass
    
    
```
